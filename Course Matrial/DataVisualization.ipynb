# Data Visualization using the tips dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe()
```

```python
tips.shape
```

```python
tips['percent'] = 100 * tips['tip']/tips['total_bill']
```

# Scatter Plot

**Plot using dataframe plotting methods**

```python
tips.plot(kind='scatter', y = 'tip', x = 'total_bill')
```

**Plots using Seaborn**

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
```

# BoxPlot

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.boxplot(y='total_bill', data=tips)
```

```python
sns.boxplot(x='sex', y='total_bill', data= tips)
```

```python
sns.boxplot(x='day', y='total_bill', data= tips)
```

**Draw a boxplot with nested grouping by two categorical variables**

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
```

```python
sns.boxplot(x="sex", y="percent", hue="day", data=tips, palette="Set3")
```

```python
sns.boxplot(x='percent', data=tips)
```

**Remove the outliers**

```python
tips = tips[ tips.percent < 60]
```

```python
sns.boxplot(x="sex", y="percent", hue="day", data=tips, palette="Set3")
```

```python
sns.boxplot(x="sex", y="percent", hue="smoker", data=tips, palette="Set3")
```

**Use swarmplot() to show the datapoint on top of the boxes**

```python
sns.boxplot(x="day", y="percent", data=tips, palette="Set3")
sns.swarmplot(x="day", y="percent", data=tips, color=".1")
```

**Use catplot**

```python
sns.catplot(x="sex", y="total_bill",
...                 hue="smoker", col="time",
...                 data=tips, kind="box",
...                 height=4, aspect=.7);
```
